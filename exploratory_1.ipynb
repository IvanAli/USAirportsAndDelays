```python
%matplotlib inline
%autosave 10
import warnings
import matplotlib.pyplot as plt
import pandas as pd
warnings.filterwarnings('ignore')
```

```python
data1 = pd.read_csv('data/On_Time_On_Time_Performance_2015_1.csv', sep=',')
data2 = pd.read_csv('data/On_Time_On_Time_Performance_2015_2.csv', sep=',')
# data4 = pd.read_csv('data/On_Time_On_Time_Performance_2015_4.csv', sep=',')
# data5 = pd.read_csv('data/On_Time_On_Time_Performance_2015_5.csv', sep=',')
# data6 = pd.read_csv('data/On_Time_On_Time_Performance_2015_6.csv', sep=',')
# data7 = pd.read_csv('data/On_Time_On_Time_Performance_2015_7.csv', sep=',')
# data8 = pd.read_csv('data/On_Time_On_Time_Performance_2015_8.csv', sep=',')
```

```python
features = ['Quarter', 'Month', 'UniqueCarrier', 'FlightNum', 'OriginAirportID',
            'DestAirportID', 'CRSDepTime', 'DepDelay', 'CRSArrTime', 'Cancelled', 'Distance']
df1 = data1.ix[:, features]
df2 = data2.ix[:, features]
# df5 = data5.ix[:, features]
# df6 = data6.ix[:, features]
# df7 = data7.ix[:, features]
# df8 = data8.ix[:, features]
# df9 = data9.ix[:, features]
```

```python
df = pd.merge(df1, df2, how='outer')
```

```python
df1.drop_duplicates(['UniqueCarrier']).head(15)
```

### Number of cancelled occurances per carrier

```python
dfgroup = df.groupby(['UniqueCarrier'])
dfgroup.describe()
```

```python
for name, group in dfgroup:
    print('For carrier', name)
    cancelled = len(group[group['Cancelled'] == 1])
    total = len(group)
    ratio = cancelled / total
    print('Number of cancelled flights:', cancelled, '\nNumber of total flights:', total,
         'Ratio:', ratio)
```

### Carriers with cancel percentage higher than 3%

```python
for name, group in dfgroup:
    cancelled = len(group[group['Cancelled'] == 1])
    total = len(group)
    ratio = cancelled / total
    if (ratio > 0.03):
        print('Carrier', name, 'has ratio', ratio);
```

### Plotting ratios

```python
x = []
xstrings = []
y = []
for name, group in dfgroup:
    cancelled = len(group[group['Cancelled'] == 1])
    total = len(group)
    ratio = cancelled / total
    x.append(len(x))
    xstrings.append(name)
    y.append(ratio)
    
# plt.plot(x, y, 'bs')
 
plt.bar(x, y, align='center', alpha=0.5)
plt.xticks(x, xstrings)
plt.ylabel('Cancellation ratio')
plt.title('Ratio (cancelled flights / total flights) for each carrier')
 
# plt.show()
```
